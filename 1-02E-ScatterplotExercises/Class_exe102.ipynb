```python
# Perform imports here:
import pandas as pd 
import numpy as np
import plotly.express as px
import plotly.graph_objects as go 
import plotly.offline as pyo
```

# Objective: Create a scatterplot of 1000 random data points.
## x-axis values should come from a normal distribution using
## np.random.randn(1000)
## y-axis values should come from a uniform distribution over [0,1) using
## np.random.rand(1000)

```python
x1=  np.random.randn(1000)
y1 = np.random.rand(1000)
# Define a data variable
d1 = [go.Scatter(x = x1, 
                 y = y1, 
                 mode = 'markers',
                 marker = {
                            'size' : 2,
                            'color': 'rgb(151,4,13)',
                            'symbol': 'pentagon',
                            'line': {'width': 2 }})]

# Define the layout

layout2 = go.Layout(title = 'Hello My second Plot',
                    xaxis = {'title': 'X-axis'},
                    yaxis = {'title': 'Y-axis'},
                    hovermode = 'closest'
)

# Create a fig from data and layout, and plot the fig
fig2 = go.Figure(data = d1, layout = layout2)
pyo.plot(fig2)
```

```python
import plotly.express as px

d2 = pd.DataFrame({'x': [i for i in x1], 'y': [i for i in y1]})

ff1 = px.scatter(d2, x = x1, y = y1)

ff1.update_layout(
    title = 'Hello My second Plot',
                    xaxis = {'title': 'X-axis'},
                    yaxis = {'title': 'Y-axis'},
                    hovermode = 'closest'
)

ff1.update_traces(
    marker = {
                            'size' : 5,
                            'color': 'rgb(151,4,13)',
                            'symbol': 'pentagon',
                            'line': {'width': 2 }}
)

ff1.show()
```
