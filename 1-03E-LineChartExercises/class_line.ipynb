```python
import pandas as pd 
import numpy as np 
import plotly.express as px
import plotly.graph_objects as go
import plotly.offline as pyo 

np.random.seed(56)
```

```python
x_a = np.linspace(0,1,100)

y_a = np.random.randn(100)
```

```python
trace0 = go.Scatter(x= x_a, y = y_a + 5, mode = 'markers', name = 'markers')

trace1 = go.Scatter(x= x_a, y = y_a, mode = 'lines', name = 'line1')

trace2 = go.Scatter(x= x_a, y = y_a -5 , mode = 'lines + markers', name = 'best')

d1 = [trace0,trace1, trace2]

layout1 = go.Layout(title = 'line Chart')

fig = go.Figure(data = d1 , layout= layout1)

pyo.plot(fig)
```

```python
d2 = pd.DataFrame({'X_a': x_a.tolist(), 'Y_a': (y_a + 5).tolist() })
d3 = pd.DataFrame({'X_a': x_a.tolist(), 'Y_a': (y_a ).tolist() })

l1 = px.line(d2, x = 'X_a', y = 'Y_a')

l2 = px.scatter(d3, x = 'X_a', y = 'Y_a')

p1 = go.Figure()

# Add both traces
for trace in l1.data:
    p1.add_trace(trace)

for trace in l2.data:
    p1.add_trace(trace)

p1.update_layout(
    title = 'Line Chart',
    xaxis = dict(title = 'x_axis'),
    yaxis = dict(title = 'y_axis')

)
p1.show()
```

```python
import plotly.express as px
import plotly.graph_objects as go

import pandas as pd

# Sample Data
df = pd.DataFrame({
    'x': [1, 2, 3, 4],
    'y_line': [2, 3, 4, 5],
    'y_scatter': [1, 4, 2, 3]
})

# Create line chart using px
fig_line = px.line(df, x='x', y='y_line')

# Create scatter chart using px
fig_scatter = px.scatter(df, x='x', y='y_scatter')

# Combine traces into one figure
fig_combined = go.Figure()

# Add both traces
for trace in fig_line.data:
    fig_combined.add_trace(trace)

for trace in fig_scatter.data:
    fig_combined.add_trace(trace)

fig_combined.update_layout(title='Line + Scatter Combined')
fig_combined.show()
```

```python
#######
# Objective: Using the file 2010YumaAZ.csv, develop a Line Chart
# that plots seven days worth of temperature data on one graph.
# You can use a for loop to assign each day to its own trace.
######

# Create a pandas DataFrame from 2010YumaAZ.csv
df = pd.read_csv('../data/2010YumaAZ.csv')
days = ['TUESDAY','WEDNESDAY','THURSDAY','FRIDAY','SATURDAY','SUNDAY','MONDAY']


df.head()
```

```python
# Use a for loop (or list comprehension to create traces for the data list)
data_f = []

for day in days:
    # What should go inside this Scatter call?
    trace = go.Scatter(x=df.LST_TIME , y= df.loc[df.DAY ==day,'T_HR_AVG'], mode = 'lines', name= day)
    data_f.append(trace)

# Define the layout


layout_final = go.Layout(title = 'line Chart for weekday temp',
                         xaxis = dict(title = 'Time'),
                         yaxis = dict(title = 'Temp'))



# Create a fig from data and layout, and plot the fig

fig_f = go.Figure(data = data_f , layout= layout_final)

pyo.plot(fig_f)
```

```python


# Use a for loop (or list comprehension to create traces for the data list)
data_ext = [go.Scatter(x= df.LST_TIME.unique(),
                       y = df.loc[df.DAY == d, 'T_HR_AVG'],
                       mode = 'lines',
                       name = d) for d in days]


# Define the layout
layout_extra = go.Layout(title = 'line Chart for weekday temp',
                         xaxis = dict(title = 'Time'),
                         yaxis = dict(title = 'Temp'))



# Create a fig from data and layout, and plot the fig

fig_extra = go.Figure(data = data_ext , layout= layout_extra)

pyo.plot(fig_extra)




# Create a fig from data and layout, and plot the fig
```
