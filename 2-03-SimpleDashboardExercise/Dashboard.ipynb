```python
#######
# Objective: build a dashboard that imports OldFaithful.csv
# from the data directory, and displays a scatterplot.
# The field names are:
# 'D' = date of recordings in month (in August),
# 'X' = duration of the current eruption in minutes (to nearest 0.1 minute),
# 'Y' = waiting time until the next eruption in minutes (to nearest minute).
######

# Create a DataFrame from the .csv file:
ds_data = pd.read_csv('../data/OldFaithful.csv', header = 0)
ds_data.head()
```

```python
# Create a Dash layout that contains a Graph component:

prac = dash.Dash()

prac.layout = html.Div(
    [dcc.Graph(
        id = 'first_plot',
        figure= {
            'data':[go.Scatter(x = ds_data.X, 
                               y = ds_data.Y,
                               mode = 'markers',
                               marker= dict(
                                   size = 6,
                                   color = 'green',
                                   symbol = 'star',
                                   line = dict(width = 1)
                               )
                               )],
            'layout': go.Layout(title = 'First exercise dashboard!',
                                xaxis= dict(title = 'x'),
                                yaxis= dict(title = 'Y'),
                                hovermode= 'closest')
        }
    ),
    dcc.Graph(
        id = 'Sec_plot',
        figure= {
            'data':[go.Scatter(x = ds_data.D.drop_duplicates(), 
                               y = ds_data.groupby('D')['Y'].mean().reset_index().Y,
                               mode = 'lines'
                               )],
            'layout': go.Layout(title = 'sec exercise dashboard!',
                                xaxis= dict(title = 'x'),
                                yaxis= dict(title = 'Y'),
                                hovermode= 'closest')
        }
    )
    ]
)

prac.run()
```
