```python
d2 = dash.Dash()

d2.layout = html.Div([
    html.Label('First Dropdown', style= dict(color = 'blue')),
    dcc.Dropdown(
        options = [
            dict(label = 'New York City', value = 'NYC'),
            dict(label = 'Washinton', value = 'WA'),
            dict(label = 'US', value = 'All')
        ], value= 'WA'
    ),
    html.Label('Section Dropdown', style= dict(color = 'Green')),
    dcc.Dropdown(
        options = [
            dict(label = 'New York City', value = 'NYC'),
            dict(label = 'Washinton', value = 'WA'),
            dict(label = 'US', value = 'All')
        ], value= ['WA', 'NYC'], multi= True
    ),
    html.Label('Slider', style= dict(color = 'red')),
    dcc.Slider(min = -5,
               max = 25,
               step = 1,
               marks = {i: i for i in range(-5,26)},
               value = 0
    ),
    html.Label('Radio Items', style= dict(color = 'yellow')),
    dcc.RadioItems(options = [
            dict(label = 'New York City', value = 'NYC'),
            dict(label = 'Washinton', value = 'WA'),
            dict(label = 'US', value = 'All')
        ], value= 'WA'
    ),
    html.Label('Check box', style= dict(color = 'gray')),
    dcc.Checklist(
    options=[
        {'label': 'New York City', 'value': 'NYC'},
        {'label': 'Washington', 'value': 'WA'},
        {'label': 'US', 'value': 'All'}
    ],
    value=['WA'],  # Initial selected values
    labelStyle={'display': 'block'}
)

])

# d2.run()
```

```python
import dash
from dash import dcc, html, Output, Input

app = dash.Dash(__name__)

options = [
    {'label': 'New York City', 'value': 'NYC'},
    {'label': 'Washington', 'value': 'WA'},
    {'label': 'US', 'value': 'All'},
     {'label': 'hv', 'value': 'haaaa'}
]

app.layout = html.Div([
    dcc.Checklist(
        id='my-checklist',
        options=options,
        value=[],
        labelStyle={'display': 'block'}
    ),
    html.Div(id='selected-values')
])

@app.callback(
    Output(component_id = 'my-checklist', component_property='value'),
    Input(component_id = 'my-checklist', component_property='value'),
    prevent_initial_call=True
)
def limit_selection(selected):
    if len(selected) > 2:
        return selected[:2]  # Trim to max 2
    return selected

# if __name__ == '__main__':
#     app.run(debug=True)

# app.run()
```
